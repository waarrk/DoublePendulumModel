```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

```python
def equationfunc(t, y):# 計算したい微分方程式を関数として作成する
    dydt = np.zeros_like(y)
    dydt[0] = y[1]
    dydt[1] = -y[0]
    return dydt
```

```python
t = np.linspace(0, 10, 100)#範囲　数
solved = scipy.integrate.solve_ivp(equationfunc, [0,20], [0,1],  t_eval=t) #関数 範囲 初期値 時間刻み幅

graph1t = solved.t
graph2y = solved.y[1,:]
```

```python
graph2t = np.linspace(0, 10, 100) #等差数列生成 #最初　最後 要素数

plt.title("scipy.integrate.solve_ivp")
plt.xlabel("t")
plt.ylabel("y")
plt.minorticks_on()
plt.grid()

plt.plot(graph1t,  graph2y, 'o',  label="solved") #微分方程式の解曲線をプロット
plt.plot(graph2t, np.cos(graph2t), '--',  label="ideal")#理想的な指数関数をプロット

plt.legend()

plt.show()
```
