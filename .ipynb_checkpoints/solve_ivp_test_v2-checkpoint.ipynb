```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
def equationfunc(t, y):# 計算したい微分方程式を関数として作成する
    dydt = y
    return dydt
```

```python
solved = scipy.integrate.solve_ivp(equationfunc, [0,10], [1]) #関数 範囲 初期値

graph1t = solved.t
graph2y = solved.y.T
```

```python
graph2t = np.linspace(0, 10, 100) #等差数列生成 #最初　最後 要素数

plt.title("scipy.integrate.solve_ivp")
plt.xlabel("t")
plt.ylabel("y")
plt.minorticks_on()
plt.grid()

plt.plot(graph1t,  graph2y, label="solved") #微分方程式の解曲線をプロット
plt.plot(graph2t, np.exp(graph2t), label="ideal")#理想的な指数関数をプロット

plt.legend()

plt.show()
```
